# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF -using the equation- of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

```python
#type your code here:
mu , sigma = 0, 1

# Creating Sample that is normally distributed
x = np.random.normal(loc=mu,scale=sigma,size= 5000)

# Computing pdf
def normal_pdf(x,mu,sigma):
    return (1 / (sigma* np.sqrt(2 * np.pi))) * np.exp((-1/2) * ((x-mu) / sigma) ** 2)

mu, sigma = stats.norm.fit(x)
points = np.linspace(min(x),max(x),5000)
pdf = normal_pdf(points,mu,sigma)

plt.hist(x,20, density=True);
plt.plot(points, pdf, color='b')
plt.show() 

print(f"Sample mean : {mu} \n Sample Variance = {sigma ** 2}")
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
#type your code here:
mu = 0
sigma = 1 

x = np.random.normal(loc=mu,scale=sigma,size= 5000)

    
mu, sigma = stats.norm.fit(x)
points = np.linspace(min(x),max(x),5000)
pdf = stats.norm.pdf(points,loc=0,scale=1)
cdf = stats.norm.cdf(points,loc=0,scale=1)


plt.hist(x,20, density=True)
plt.plot(points, pdf, color='b')
plt.plot(points,cdf,color='y')

plt.show() 
print(f"Sample mean : {np.mean(x)} \n Sample Variance = {np.var(x)}")
```

C) Repeat (a) for an exponential random variable with $\lambda = 1$ using [numpy.random.exponential](https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html), and [scipy.stats.expon](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon)

```python
#type your code here:
lamda=1
x_exp = np.random.exponential(scale=lamda,size= 5000)

points_3 = np.linspace(min(x_exp),max(x_exp),50000)

pdf_exp = stats.expon.pdf(points_3,scale=lamda)

plt.hist(x_exp,30, density=True);
plt.plot(points_3, pdf_exp, color='b')
plt.show() 

print(f"Sample mean : {abs(np.mean(x_exp))} \n Sample Variance = {abs(np.var(x_exp))}")
```

```python
#type your code here:

lamda = 1
x_exp = np.random.exponential(scale=lamda,size= 5000)

count , points_3,_ = plt.hist(x_exp,30,density = True,label='Historam',color='b')
pdf_exp = stats.expon.pdf(points_3,scale=lamda)
cdf_exp = stats.expon.cdf(points_3,scale=lamda)

# plt.hist(x_exp,30, density=True);
plt.plot(points_3, pdf_exp, color='r',label='pdf')
plt.plot(points_3, cdf_exp, color='orange',label='cdf')
plt.legend()
plt.show() 

print(f"Sample mean : {abs(np.mean(x_exp))} \n Sample Variance = {abs(np.var(x_exp))}")
```

e) Repeat for an binomial random variable with $n = 5$ and $p = 0.5$

```python
#type your code here:
n=5
p = 0.5
x_bio = np.random.binomial(n,p,5000)

count, bins,_= plt.hist(x_bio,5, density=True, label='histogram')

x_vals = list(range(6))
distance = [stats.binom.pmf(vals,n,p) for vals in x_vals]

plt.bar(x_vals,distance, label='PMF', color='b', width=0.04)

plt.plot()

print(f"Sample mean : {np.mean(x_bio)} \n Sample Variance = {np.var(x_bio)}")
```

```python
#type your code here:
n=5
p=0.5
x_bio = np.random.binomial(n,p,5000)

x_pmf = list(range(6))

distance = [stats.binom.pmf(vals,n,p) for vals in x_pmf]

lin=np.arange(0,5,0.001)

cdf = stats.binom.cdf(lin,5,0.5)

plt.plot(lin,cdf, label='CDF', color='b')

plt.bar(x_pmf,distance, label='PMF', color='red', width=0.04)


plt.plot()


```
